3.CNN Feature Extraction with Filters and Pooling

Task 1: Implement Edge Detection Using Convolution

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

def edge_detection_sobel(image_path):
    """
    Applies Sobel filter for edge detection on a grayscale image.

    Args:
        image_path (str): Path to the image file.
    """
    try:
        # Load the image in grayscale
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        if img is None:
            raise FileNotFoundError(f"Could not open or find the image: {image_path}")

        # Define Sobel filters
        sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
        sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

        # Apply Sobel filter in x-direction
        sobelx = cv2.filter2D(img, cv2.CV_64F, sobel_x)
        sobelx = np.uint8(np.absolute(sobelx))

        # Apply Sobel filter in y-direction
        sobely = cv2.filter2D(img, cv2.CV_64F, sobel_y)
        sobely = np.uint8(np.absolute(sobely))

        # Display the images
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        plt.imshow(img, cmap='gray')
        plt.title('Original Image')

        plt.subplot(1, 3, 2)
        plt.imshow(sobelx, cmap='gray')
        plt.title('Sobel X')

        plt.subplot(1, 3, 3)
        plt.imshow(sobely, cmap='gray')
        plt.title('Sobel Y')

        plt.show()

    except FileNotFoundError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage (replace 'your_image.jpg' with your image file path)
image_path = '/content/Tig.jpg' # example image path.
edge_detection_sobel(image_path)
```

Task 2: Implement Max Pooling and Average Pooling

```
import tensorflow as tf
import numpy as np

def pooling_demo():
    """
    Demonstrates Max Pooling and Average Pooling on a random 4x4 matrix.
    """

    # Create a random 4x4 matrix as input
    input_matrix = np.random.randint(0, 10, size=(1, 4, 4, 1))  # (batch_size, height, width, channels)
    input_tensor = tf.constant(input_matrix, dtype=tf.float32)

    # Max Pooling
    max_pool = tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='valid')(input_tensor)

    # Average Pooling
    avg_pool = tf.keras.layers.AveragePooling2D(pool_size=(2, 2), strides=(2, 2), padding='valid')(input_tensor)

    # Print the results
    print("Original Matrix:")
    print(input_matrix[0, :, :, 0])  # Remove batch and channel dimensions for printing

    print("\nMax Pooled Matrix:")
    print(max_pool.numpy()[0, :, :, 0])

    print("\nAverage Pooled Matrix:")
    print(avg_pool.numpy()[0, :, :, 0])

# Run the pooling demo
pooling_demo()
```
